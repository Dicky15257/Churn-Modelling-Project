##**Feature Engineering**

```python
df = pd.get_dummies(df, columns=['Gender'], drop_first=True)
df = pd.get_dummies(df, columns=['Geography'], drop_first=False)
```

```python
df.tail()
```

```python
X = df[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
        'IsActiveMember', 'EstimatedSalary', 'Gender_Male', 'Geography_France', 'Geography_Germany', 'Geography_Spain']]
y = df['Exited']
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

## Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

## Ensemble

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import confusion_matrix
def classification_eval(aktual, prediksi, nama):
    cm = confusion_matrix(aktual, prediksi)
    tp = cm[1][1]
    tn = cm[0][0]
    fp = cm[0][1]
    fn = cm[1][0]
    
    accuracy = round((tp+tn) / (tp+tn+fp+fn) * 100, 2)
    precision = round((tp) / (tp+fp) * 100, 2)
    recall = round((tp) / (tp+fn) * 100, 2)
    f1_score = 2*(precision*recall)/(precision+recall)
    print("Confusion Matrix:")
    print(cm)
    print()
    print('Accuracy   : {:.2f}'.format(accuracy) , '%')
    print('Precision  : {:.2f}'.format(precision), '%')
    print('Recall     : {:.2f}'.format(recall), '%')
    print('F1-Score   : {:.2f}'.format(f1_score), '%')
```

##KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
#KNN training model
KNN = KNeighborsClassifier()
KNN.fit(X_train, y_train)
```

```python
y_train_predKNN = KNN.predict(X_train)
y_test_predKNN = KNN.predict(X_test)
```

```python
#Evaluasi pada training model
classification_eval(y_train, y_train_predKNN, 'KNN Training Perfomance')
```

```python
#Evaluasi pada Testing Model
classification_eval(y_test, y_test_predKNN, 'KNN test Perfomance')
```

##Decision Tree

```python
DecTree = DecisionTreeClassifier()
DecTree.fit(X_train, y_train)
```

```python
y_train_predDT = DecTree.predict(X_train)
y_test_predDT = DecTree.predict(X_test)
```

```python
#Evaluasi pada training model
classification_eval(y_train, y_train_predDT, 'Decision Tree Training Perfomance')
```

```python
#Evaluasi pada testing model
classification_eval(y_test, y_test_predDT, 'Decision Tree Testing Perfomance')
```

## Random Forest

```python
RanFor = RandomForestClassifier(max_depth = 9)
RanFor.fit(X_train, y_train)
```

```python
y_train_predRF = RanFor.predict(X_train)
y_test_predRF = RanFor.predict(X_test)
```

```python
#Evaluasi pada training model
classification_eval(y_train, y_train_predRF, 'Random Forest Training Perfomance')
```

## Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier

clf = AdaBoostClassifier(n_estimators=50, learning_rate=1, random_state=42)
clf.fit(X_train, y_train)

y_train_predAB = clf.predict(X_train)
y_pred = clf.predict(X_test)
classification_eval(y_train, y_train_predAB, "Ada Boost Training Perf")
```

```python
classification_eval(y_test, y_pred, "Ada Boost Testing Perf")
```

## GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(n_estimators=50, random_state=42)
clf.fit(X_train, y_train)
y_train_predGB = clf.predict(X_train)
y_pred = clf.predict(X_test)
classification_eval(y_train, y_train_predGB, "Gradient Boost Training Perf")
```

```python
classification_eval(y_test, y_pred, "Gradient Boost Testing Perf")
```

## XGBoost

```python
import xgboost as xgb

clf = xgb.XGBClassifier()
clf.fit(X_train, y_train)

y_train_predXGB = clf.predict(X_train)
y_pred = clf.predict(X_test)
classification_eval(y_train, y_train_predXGB, 'XGB Training Perf.')
```

```python
classification_eval(y_test, y_pred, 'XGB Testing Perf.')
```

## Stacking

```python
from sklearn.ensemble import StackingClassifier
estimators = [
     ('1', AdaBoostClassifier()),
     ('2', GradientBoostingClassifier()),
     ('3', RandomForestClassifier()),
     ('4', DecisionTreeClassifier()),
     ('5', KNeighborsClassifier()),
     ]
clf = StackingClassifier(
      estimators=estimators, final_estimator=xgb.XGBClassifier()
      )
clf.fit(X_train, y_train)
y_train_predSt = clf.predict(X_train)
y_pred = clf.predict(X_test)
classification_eval(y_train, y_train_predSt, 'Stacking Training Performance')
```

```python
classification_eval(y_test, y_pred, 'Stacking Testing Performance')
```

## Hyperparameter Tuning

Grid Search for Random Forest

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestClassifier()

param_grid = {'n_estimators': [10, 25, 50], 
              'max_depth': [10, 50, None], 
              'bootstrap': [True, False]}


grid_search_forest = GridSearchCV(forest, param_grid, cv=10, scoring='f1')
grid_search_forest.fit(X_train, y_train)
```

```python
cvres = grid_search_forest.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.round(mean_score, 2), params)
```

```python
grid_search_forest.best_params_
```

```python
gs_result = grid_search_forest.best_estimator_
y_pred = gs_result.predict(X_test)
classification_eval(y_test, y_pred, 'Random Forest Perfomance After Hyper-Tuning')
```

## Randomized Search

#Randomized Search for Random Forest


```python
from sklearn.model_selection import RandomizedSearchCV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 20, stop = 200, num = 5)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(1, 45, num = 3)]
# Minimum number of samples required to split a node
min_samples_split = [5, 10]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split}

random_grid
```

```python
rf_random = RandomizedSearchCV(estimator = forest, param_distributions = random_grid, n_iter = 10, cv = 10,
                               random_state=42, n_jobs = -1, scoring='f1')
# Fit the random search model
rf_random.fit(X_train, y_train)
```

```python
cvres = rf_random.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.round(mean_score, 2), params)
```

```python
rf_random.best_params_
```

```python
rs_res = rf_random.best_estimator_
y_pred = rs_res.predict(X_test)
classification_eval(y_test, y_pred, 'Random Forest Performance After Randomized Search')
```
